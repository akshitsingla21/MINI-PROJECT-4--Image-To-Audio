```python
import pytesseract
from PIL import Image
import gtts
```

#adding file

```python
img = Image.open('/content/Picture1.png')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
tts = gtts.gTTS(result)
tts.save("pngspeak.png")
```

#playing the audio

```python
from IPython.display import Audio
wn = Audio('/content/pngspeak.mp3', autoplay=True)
display(wn)
```
